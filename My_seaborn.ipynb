### sns.lineplot

```python
x = np.linspace(0, 10, 100)
y = x ** 2
sns.lineplot(x=x, y=y)
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='sex', size='size')
```

## Figure-level functions

### sns.relplot

```python
sns.relplot?
```

```python
sns.relplot(kind='scatter', data=tips, x='total_bill', y='tip', col='sex', hue='smoker') #col='sex'คือแบ่งกราฟแยกตามเพศ ทำrowได้เช่นกัน
```

```python
sns.relplot(kind='scatter', data=tips, x='total_bill', y='tip', row='sex', col='day', hue='smoker')
```

```python
sns.relplot(
    kind='scatter', data=tips, 
    x='total_bill', y='tip', row='sex', 
    col='day', hue='smoker', 
    height=6, aspect=0.8) #aspect=กว้าง =1คือเท่าความสูง
# height - Height (in inches) of each facet.
# aspect - Aspect ratio of each facet, so that aspect * height gives the width of each facet in inches.
```

### sns.jointplot

```python
sns.jointplot(kind='scatter', data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(kind='hex', data=tips, x='total_bill', y='tip')
```

### sns.lmplot

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')#มีเส้นregressionให้ด้วย
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', col='time')
```

# Distribution Plots

## Axes-level Functions

### sns.histplot

```python
sns.histplot(data=tips, x='total_bill')
```

```python
sns.histplot(data=tips, x='total_bill', stat='percent')#default=count
# stat options: count, frequency, probability, percent, density
```

```python
sns.histplot(data=tips['total_bill'], bins=20)
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, hue='sex')
```

### sns.kdeplot

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='day')
```

### sns.ecdfplot
ECDF = empirical cumulative distribution functions.

An ECDF represents the proportion or count of observations falling below each unique value in a dataset.

```python
sns.ecdfplot(data=tips, x='tip')
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex')
```

# Usage with matplotlib

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# scatterplot() returns matplotlib.axes.Axes object.
ax = sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
ax.set_title('Tip behavior of Male vs Female')
ax.set_ylabel('Tip ($)')
ax.set_xlabel('Total Bill ($)')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', ax=axes[0])
sns.countplot(data=tips, x='sex', ax=axes[1])
```
